# Volume 4: Introduction to Initial Value Problems and Boundary Value Problems

```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

### Problem 1
Use solve_ivp to solve for $y$ in the equation $y'' - y = sin(t)$ with initial conditions $y(0)= -\frac{1}{2}$, $y'(0) = 0$ and plot your solution on the interval $[0,5]$. Compare this to the analytic solution $y=-\frac{1}{2}(e^{-t}+sin(t))$. 

Note: Using max_step = 0.1 with give you a smoother graph here

```python
#element one: the ODE
def ode(t,y):
    return np.array([y[1],np.sin(t) + y[0]])
    #return np.array([y[1],np.sin(t)-3*y[0]])

#element two: the time domain
t_span = (0,5)

#element three: the initial conditions
y0 = np.array([-1/2,0])

#solve the system
sol = solve_ivp(ode,t_span,y0,max_step = .1)

#plot
plt.plot(sol.t,sol.y[0])
plt.title("IVP Solution")
plt.show()
t = np.linspace(0,5,75)
plt.plot(t,(-1/2)*(np.exp(-t) + np.sin(t)),'orange')
plt.title("Analytic Solution")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.show()
```

### Problem 2

Use solve_bvp to solve for $y$ in the equation $y'+y''=-\frac{1}{4}e^{-0.5t}+sin(t)-cos(t)$ with boundary conditions $y(0)=6$ and $y'(5)=-0.324705$ and plot your solution on the interval $[0,5]$. Use an initial guess of all ones. Compare this to the analytic solution $y =e^{-0.5t}-sin(t)+5$. 

```python
#element one: The ODE
def ode(t,y):
    return np.array([y[1],(-1/4)*np.exp(-.5*t) + np.sin(t) - np.cos(t) - y[1]])

#element two: Boundary condition
def bc(ya,yb):
    return np.array([ya[0]-6,yb[1]+.324705])

#element three: time domain
t_steps = 100
t = np.linspace(0,5,t_steps)

#element four: the initial guess
y0 = np.ones((2,t_steps))

#solve
sol = solve_bvp(ode, bc, t, y0)
plt.plot(sol.x, sol.y[0])
plt.title("BVP Solution")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.show()

t = np.linspace(0,5,100)
plt.plot(t,np.exp(-.5*t) - np.sin(t) + 5,'orange')
plt.title("Analytic Solution")
plt.show()
```

### Problem 3

Use solve_bvp to solve for $y$ in the equation $y''=(1-y')*10y$ with boundary conditions $y(0)=-1$ and $y(1)=\frac{3}{2}$ and plot your solution on the interval $[0,1]$. Use an initial guess of 10. Compare this to the the same solution using an initial guess of 9.99. 

```python
#element one: The ODE
def ode(t,y):
    return np.array([y[1],(1-y[1])*10*y[0]])

#element two: Boundary condition
def bc(ya,yb):
    return np.array([ya[0]+1,yb[0]-1.5])

#element three: time domain
t_steps = 50
t = np.linspace(0,1,t_steps)

#element four: the initial guess
y0 = np.ones((2,t_steps))
y1 = y0 + 9
y2 = y0 + 8.99

#solve
sol = solve_bvp(ode, bc, t, y1)
sol2 = solve_bvp(ode,bc,t,y2)
plt.plot(sol.x, sol.y[0],label = "Intial Guess of 10")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.plot(sol2.x,sol2.y[0], label = "Initial Guess of 9.99")
plt.legend()
plt.show()
```

### Problem 4
Use solve_ivp to solve the Four-Wing Attractor as described in equations (1.2), (1.3), and (1.4) where $a=0.2$, $b = 0.01$, and $c = -0.4$. Try this with 3 different initial values and plot (in three dimensions) the 3 corresponding graphs.

```python
#element one: ODE
def ode(t,x):
    return np.array([.2*x[0] + x[1]*x[2],.01*x[0] -.4*x[1] - x[0]*x[2],-x[2]-x[0]*x[1]])

#element two: the time domain
t_span = (0,200)

#element three: the initial conditions
y0 = np.array([1,1,1])
y1 = np.array([2,2,2])
y2 = np.array([7,7,7])

#solve the system
sol0 = solve_ivp(ode,t_span,y0,max_step = .1)
sol1 = solve_ivp(ode,t_span,y1,max_step = .1)
sol2 = solve_ivp(ode,t_span,y2,max_step = .1)

#create figure and axes
fig = plt.figure()
ax = plt.axes(projection = '3d')

#plot for initial point of [1,1,1]
ax.plot(sol0.y[0],sol0.y[1],sol0.y[2])
plt.title("Initial value of [1,1,1]")
plt.show()

#plot for initial point of [2,2,2]
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot(sol1.y[0],sol1.y[1],sol1.y[2])
plt.title("Initial value of [2,2,2]")
plt.show()

#plot for initial point of [7,7,7]
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot(sol2.y[0],sol2.y[1],sol2.y[2])
plt.title("Initial value of [7,7,7]")
plt.show()
```
