```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    crimes = pd.read_csv("crime_data.csv", index_col = "Year")
    crimes.drop(["Population"], axis = 1).plot(linewidth = 1, title = "All Crimes by Year")
    crimes.plot(y = ['Violent', 'Vehicle Theft', 'Aggravated Assault', 'Robbery', 'Murder', 'Forcible Rape'], subplots = True, layout = (2,3), sharey = True, title = "Plots of the Least 6 Committed Crimes by Year")
    crimes.plot(kind = 'hist', y = ['Property'], bins = 15, legend = False, title = "Frequency of Property Crimes by Year")
```

```python
prob1()
```

```python
 def prob2():
    """
    Use a visualization for each of the following to identify if trends exists 
    between Forcible Rape and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Plot each visualization.
    Return a tuple of booleans. Each entry should identify whether that 
    element correlates or not.
    """
    crimes = pd.read_csv("crime_data.csv", index_col = "Year")
    #plotting Forcible rapes and violent crimes as lines against the index
    crimes.plot(y = ["Forcible Rape", "Violent"], title = "Amount of Forcible Rapes and Violent Crimes Per Year")
    #a scatter plot of Forcible rapes against violent crimes to check correlation
    crimes.plot(kind = "scatter", x = "Forcible Rape", y = "Violent", alpha = .8, xlim = (0, max(crimes['Forcible Rape']) + 1),ylim = (0, max(crimes['Burglary']) + 1), title = "Correlation of Rapes and Violent Crimes")
    
    #plotting Forcible rapes and Burglaries as lines against the index
    crimes.plot(y = ["Forcible Rape", "Burglary"], title = "Amount of Forcible Rapes and Burglaries Per Year")
    #a scatter plot of Forcible rapes against Burglaries to check correlation
    crimes.plot(kind = "scatter", x = "Forcible Rape", y = "Burglary", alpha = .8, xlim = (0, max(crimes['Forcible Rape']) + 1),ylim = (0, max(crimes['Burglary']) + 1),title = "Correlation of Forcible Rapes and Burglaries")
    
    #plotting Forcible rapes and Aggravated Assaults as lines against the index
    crimes.plot(y = ["Forcible Rape", "Aggravated Assault"], title = "Amount of Forcible Rapes and Aggravated Assaults Per Year")
    #a scatter plot of Forcible rapes against Aggravated Assaults to check correlation
    crimes.plot(kind = "scatter", x = "Forcible Rape", y = "Aggravated Assault", alpha = .8, xlim = (0, max(crimes['Forcible Rape']) + 1),ylim = (0, max(crimes['Aggravated Assault']) + 1), title = "Correlation of Forcible Rapes and Aggravated Assaults")
    
    #In an analysis of the graphs, we see that there is a correlation with Forcible Rape and Violent Crimes 
    #as well as Aggravated Assaults. There is not, however, a correlation of Forcible Rapes with Burglary.
    
    return (True,False,True)
    #raise NotImplementedError("Problem 2 Incomplete")
```

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    crimes = pd.read_csv("crime_data.csv", index_col = "Year")
    #Distrubition of Burglary, Violent Crimes, and Vehicle Theft
    crimes.plot(kind = "box", y = ['Burglary', "Violent", "Vehicle Theft"], title = "Distrubition of Burglary, Violent Crimes, and Vehicle Theft")
    #Distribution of Vehicle Thefts against the values of Robbery
    crimes.plot(kind="hexbin", x = "Vehicle Theft", y = "Robbery", gridsize = 15,sharex = False,title = "Distribution of Vehicle Thefts against the values of Robbery")
    #raise NotImplementedError("Problem 3 Incomplete")
```
